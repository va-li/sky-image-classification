```python
# dataset labels, zip export from cvat
cvat_labels_imagenet_file_path = Path('/home/vbauer/MEGA/Master/Data Science/2024 WS/Applied Deep Learning/sky-image-classification/sky-image-classification.zip')
# where to store the labels
dataset_path = Path('/home/vbauer/MEGA/Master/Data Science/2024 WS/Applied Deep Learning/sky-image-classification/data/')
# the file defining the labels of an image: image path relative to the dataset path, label number
image_numberlabels_file_path = dataset_path / 'default.txt'
# the file defining the labels names: line number corresponds to the label number, first line is label 0
numberlabel_names_file_path = dataset_path / 'synsets.txt'
```

```python
# unzip the cvat data to the dataset path
zipfile.ZipFile(cvat_labels_imagenet_file_path).extractall(dataset_path)

assert image_numberlabels_file_path.exists(), f'{image_numberlabels_file_path} does not exist'
assert numberlabel_names_file_path.exists(), f'{numberlabel_names_file_path} does not exist'
```

```python
# read the label names
with open(numberlabel_names_file_path, 'r') as f:
    # line number corresponds to the label number, first line is label 0
    label_names = { i: name.strip() for i, name in enumerate(f.readlines()) }

# read the image labels, can be multiple labels per image separated by spaces
with open(image_numberlabels_file_path, 'r') as f:
    # image path relative to the dataset path, label numbers, can be multiple separated by spaces
    image_labels = { line.split()[0]: [int(label) for label in line.split()[1:]] for line in f.readlines() }
```

```python
print(f'Number of images: {len(image_labels)}')
print(f'Number of labeled images: {sum([ len(labels) > 0 for _, labels in image_labels.items() ])}')
print(f'Number of unlabeled images: {sum([ len(labels) == 0 for _, labels in image_labels.items() ])}')
```

```python
def display_image_with_labels(dataset_path: Path, image_files: str | List[str], image_labels: pd.DataFrame) -> None:
    image_files = [ image_files ] if isinstance(image_files, str) else image_files
    images = []
    
    for image_file in image_files:
        image_path = dataset_path / image_file
        image = hv.RGB.load_image(image_path)
        labels = image_labels.loc[image_file].to_numpy()
        labels_str = ', '.join([ label for i, label in enumerate(label_names.values()) if labels[i] == 1 ])
        image.opts(
            title=labels_str,
            xaxis=None, yaxis=None)
        images.append(image)
    
    return hv.Layout(images)

def display_image_with_labels_matplotlib(dataset_path: Path, image_files: str | List[str], image_labels: pd.DataFrame, figsize_height:int=4) -> None:
    image_files = [ image_files ] if isinstance(image_files, str) else image_files
    
    fig, axs = plt.subplots(1, len(image_files), figsize=(len(image_files) * figsize_height, figsize_height))
    
    for image_file, ax in zip(image_files, axs):
        image_path = dataset_path / image_file
        image = io.imread(image_path)
        labels = image_labels.loc[image_file].to_numpy()
        labels_str = ', '.join([ label for i, label in enumerate(label_names.values()) if labels[i] == 1 ])
        
        ax.imshow(image)
        ax.axis('off')
        ax.set_title(labels_str)
        
    plt.tight_layout()
    plt.show()
    
def display_sample_matplotlib(dataset: SkyImageMultiLabelDataset, indices: List[int], figsize_height:int=4) -> None:
    
    fig, axs = plt.subplots(1, len(indices), figsize=(len(indices) * figsize_height, figsize_height))
    
    for idx, ax in zip(indices, axs):
        sample = dataset[idx]
        image = sample['image']
        sample_label_names = sample['label_names']
        labels_str = ', '.join(sample_label_names)
        
        ax.imshow(image)
        ax.axis('off')
        ax.set_title(labels_str)
        
    plt.tight_layout()
    plt.show()
        
```
