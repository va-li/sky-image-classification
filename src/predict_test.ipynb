```python
def predict_image_label_scores(model: nn.Module, transform, image_path: str | Path) -> np.ndarray:
    model.eval()
    model.to('cpu')
    image = io.imread(image_path)
    
    image = transform(image)
    with torch.no_grad():
        prediction = model(image.unsqueeze(0)).squeeze()
    return prediction.numpy()
```

```python
from dataset import SkyImageMultiLabelDataset
data_dir = Path('/home/vbauer/MEGA/Master/Data Science/2024 WS/Applied Deep Learning/sky-image-classification/data')
ds = SkyImageMultiLabelDataset(data_dir, transform=transform)
```

```python
# show the image and the predicted_label scores ordered by highest score

def plot_image_and_predicted_label_scores(image_path: Path, label_scores: np.ndarray, label_names: List[str]):
    if len(label_scores) != len(label_names):
        raise ValueError('label_scores and label_names must have the same length')
    fig, ax = plt.subplots()
    image = io.imread(image_path)
    ax.imshow(image)
    ax.axis('off')
    plt.show()
    
    sorted_label_scores = sorted(zip(label_scores, label_names), key=lambda x: x[0], reverse=True)
    for score, label in sorted_label_scores:
        print(f'{label}: {score:.2f}')
```
